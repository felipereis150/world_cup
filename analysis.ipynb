```python
worldCup = df_results[df_results['tournament'] == 'FIFA World Cup']
worldCup['year'] = worldCup['date'].dt.year
```

```python
worldCup
```

```python
worldCup_mock = worldCup.melt(id_vars=['year', 'date', 'home_team', 'away_team', 'tournament', 'city', 'country', 'neutral'], value_vars=['home_score', 'away_score'], var_name='score_type', value_name='score')
worldCup_mock
```

```python
worldCup_mock2 = worldCup_mock.melt(id_vars=['year', 'date', 'tournament', 'city', 'country', 'neutral', 'score_type', 'score'], value_vars=['home_team', 'away_team'], var_name='team_type', value_name='team')
```

```python
total_score_by_team = worldCup_mock2.groupby(['year', 'team'])['score'].sum().reset_index().sort_values(by='year', ascending=True)
total_score_by_team.head(10)
```

```python
years_with_most_score = worldCup_mock2.groupby(['year'])['score'].sum().reset_index().sort_values(by='year', ascending=True)
years_with_most_score.head(10)
```

```python
total_score_by_team['cum_score'] = total_score_by_team.groupby('team')['score'].cumsum()
total_score_by_team
```

```python
countries = df_results.country.unique()

import csv
import requests

# Set the base URL for the API
api_url = 'https://countryflagsapi.com/png'

# Open the CSV file for writing
with open('flags.csv', 'w') as csvfile:
    # Create a CSV writer object
    writer = csv.writer(csvfile)

    # Write the header row
    writer.writerow(['country', 'flag_URL'])

    # Loop through the countries
    for country in countries:
        # Build the URL for the flag image
        flag_url = f'{api_url}/{country}'

        # Write the country name and flag URL to the CSV file
        writer.writerow([country, flag_url])
flags = pd.read_csv('data/flags.csv', encoding='ISO-8859-1', engine='python')
flags
```

```python
total_score_by_team = total_score_by_team.merge(flags, left_on='team', right_on='country', how='left')
```
